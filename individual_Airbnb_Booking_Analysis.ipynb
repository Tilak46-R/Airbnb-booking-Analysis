# **REMOVAL OF STOPWORDS**

```python
from wordcloud import WordCloud
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
print(stopwords.words('french'))
```

```python
print(stopwords.words('greek'))
```

```python
def stopword_removal(data):
  stop = stopwords.words('english')
  data.name = data.name.str.replace(",","").str.lower()
```

```python
Most_repeated_words = stopword_removal(airbnb_data)
Most_repeated_words = " ".join(str(each) for each in airbnb_data.neighbourhood)
```

```python
wordcloud = WordCloud(width = 1980 , height = 1080, colormap = 'Paired', max_words = 100, background_color = 'black').generate(Most_repeated_words)
plt.figure(figsize = (10,15))
plt.figure(figsize = (15,20))
plt.imshow(wordcloud,interpolation = 'sinc')
plt.axis( "off")
plt.show()
```
