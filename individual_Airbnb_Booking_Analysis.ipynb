# **REMOVAL OF STOPWORDS**

```python
from wordcloud import WordCloud
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
def stopword_removal(data):
  stop = stopwords.words('english')
  data.name = data.name.str.replace(",","").str.lower()
```

```python
Most_repeated_words = stopword_removal(airbnb_data)
Most_repeated_words = " ".join(str(each) for each in airbnb_data.neighbourhood)
```

```python
wordcloud = WordCloud(width = 1980 , height = 1080, colormap = 'copper', max_words = 100, background_color = 'white').generate(Most_repeated_words)
plt.figure(figsize = (10,15))
plt.figure(figsize = (15,20))
plt.imshow(wordcloud,interpolation = 'sinc')
plt.axis( "off")
plt.show()
```

# **ROOM TYPES BASED ON DIFFERENT LOCATIONS**

Latitude and longitude forms a grid system which helps to identify the exact or absolute, locations on the surface of earth.latitude and longitude can be utilised to identify specific locations,which can also be helpful in identifying landmarks. On that bassis we can create a scatter plot map of density of properties and their room types.

so,lets find the latitude and longitude!

```python
airbnb_data[['latitude','longitude']]
```

```python
plt.figure(figsize = (20,10))
sca = sns.scatterplot(x = airbnb_data.longitude,y = airbnb_data.latitude, hue = airbnb_data.room_type)
sca.set_title('Density of rooms',fontsize = 15)
sca.set_xlabel('longitude',fontsize = 15)
sca.set_ylabel('latitude',fontsize = 15)
plt.show()
```

## **Room Type vs Neighbourhood Group**

```python
total_group = airbnb_data.groupby(['neighbourhood_group']).room_type.value_counts()
```

```python
total_group
```

```python
plt.figure(figsize = (20,15))
plt.title("Room Type vs Neighbourhood Group", fontsize = 25,y = 0.96)
sns.countplot(x = airbnb_data.neighbourhood_group, hue = airbnb_data.room_type, palette = 'Spectral')
plt.show()
```

## **CORRELATION PLOT TO UNDERSTAND THE RELATIONSHIP BETWEEN VARIABLES**

```python
corr = airbnb_data.corr(method = 'kendall')
plt.figure(figsize = (15,12))
plt.title("correlation between variables")
sns.heatmap(corr, annot = True, cmap = 'Spectral', linewidths = 0.1)
plt.show()
```

## **NEIGHBOURHOOD VS NUMBER_OF_REVIEWS**

```python
availability = airbnb_data.groupby("neighbourhood")['number_of_reviews'].sum().sort_values(ascending = False)
```

```python
most_reviewed = availability.head(10)
```

```python
most_reviewed
```

```python
plt.figure(figsize = (15,15))
plt.title("Neighbourhood vs Review")
sns.barplot(data = most_reviewed, palette = 'Spectral')
plt.xlabel('neighbourhood')
plt.ylabel("num_of_reviews")
plt.show()
```
