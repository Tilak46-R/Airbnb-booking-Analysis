*  ANOVA, Analysis Of Variance, which is a computational method to divide variations in an observations set into different components.
*  Correlation and correlation methods

```python
from wordcloud import WordCloud
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
def stopword_removal(data):
  stop = stopwords.words('english')
  data.name = data.name.str.replace(",","").str.lower()
```

## **CORRELATION PLOT TO UNDERSTAND THE RELATIONSHIP BETWEEN VARIABLES**

## **Categorisation of prices**

```python
airbnb_data['price_category']=airbnb_data.apply((lambda x: 'cheap' if x['price']<100 else 'affordable' if x['price']>100 and x['price']<=400 else 'Expensive' ),axis=1)
airbnb_data.head()
```

```python
Price_category = airbnb_data.groupby('price_category')['id'].count().reset_index(name = 'Total')
Price_category
```

Prices are categorised into expensive, afforadable, and cheap where majority of the rooms fall under cheaper price category.

```python
colors = ['pink','orange','grey']

explode = (0.05,0.05,0.05)

Price_category.groupby(['price_category']).sum().plot(
    kind = 'pie' , y = 'Total',autopct='%1.0f%%',
    colors= colors, explode=explode
)
```

```python
price_category_counts = airbnb_data.groupby(['neighbourhood_group','price_category'])['id'].count().reset_index(name = 'Total')
price_category_counts
```

```python
plt.figure(figsize = (15,12))
plot = sns.barplot(y = 'Total', x = 'neighbourhood_group', hue ='price_category',data = price_category_counts)
plot.set_title("Relationship between price category and neighbourhood group")
plot.set_xlabel("Neighbourhood_group")
plot.set_ylabel("price_category")
plt.show()
```
